```r
data <- read.csv(file="tmax_data.csv", header=TRUE, sep=",")
head(data)
```

```r
mod <- glm(Deaths~I(Tmax^2), data=data, family=poisson)
```

```r
summary(mod)
```

```r
par(mfrow=c(2, 2))
plot(mod)
```

```r
data$gen_deaths = 1.602e-05*I(data$Tmax^2) + 3.123
```

```r
data
```

```r
parabola <- data.frame("Tmax" = c(30, 65, 95) , "Deaths" = c(40, 28, 45))
```

```r
mod2 <- glm(deaths ~ I(tmax^2), data=parabola, family="poisson")
```

```r
summary(mod2)
```

```r
data$gen_deaths2 <- predict(mod2, data$Tmax, type="response")
```
