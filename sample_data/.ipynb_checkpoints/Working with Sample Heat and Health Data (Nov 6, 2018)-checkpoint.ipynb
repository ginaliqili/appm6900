```r
# this data shows the date, tmax, and my manually created "death count" for a particular station in VA
data <- read.csv(file="tmax_data.csv", header=TRUE, sep=",")
head(data)
```

```r
# I made a parabola here that is representative of what the exposure response curve should look like. I modeled this and then used the model formula to generate deaths
parabola <- data.frame("Tmax" = c(30, 45, 65, 80, 95), "Deaths" = c(40, 15, 10, 20, 45))
mod2 <- lm(Deaths~Tmax + I(Tmax^2), data=parabola)
```

```r
plot(parabola)
lines(fitted(mod2) ~ Tmax, data = parabola, col = "red")
```

```r
data$gen_deaths <- predict(mod2, data, type="response")
```
