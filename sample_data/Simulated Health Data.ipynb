```r
library(ggplot2)
set.seed(12)

#construct predictors 
n = 100

d <- data.frame(
   x1=rnorm(n, mean = 45, sd = 15),
   x2 = rnorm(n, mean = 60, sd=10),
   x3=sample(c('m','t','w', 'th', 'f', 'sa', 'su'),size=n,replace=TRUE),
   x4=sample(c(1990:2016),size=n,replace=TRUE),
 stringsAsFactors=F)

d$mu = with(d, exp(log(x1 + 0.5*x1^2 + 0.75*x1^3) + as.integer(as.factor(x3)) - as.integer(as.factor(x4))))

d$y = rpois(n, d$mu); #head(d$y)
with(d, plot(x1, y))
```
